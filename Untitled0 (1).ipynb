```python
import numpy as np
import pandas as pd
from math import sin, cos
from scipy.optimize import least_squares, minimize, Bounds
import matplotlib.pyplot as plt

# Parametric model function
def model_xy(t, theta_rad, M, X):
    # t can be array
    e = np.exp(M * np.abs(t))
    s = np.sin(0.3*t)
    x = t*np.cos(theta_rad) - e * s * np.sin(theta_rad) + X
    y = 42.0 + t*np.sin(theta_rad) + e * s * np.cos(theta_rad)
    return x, y

# Euclidean residuals for a set of paired t_i and observations
def residuals_for_params_with_t(params, t, x_obs, y_obs):
    theta_deg, M, X = params
    theta_rad = np.deg2rad(theta_deg)
    x_pred, y_pred = model_xy(t, theta_rad, M, X)
    # use L2 residuals for optimization (we'll compute L1 score separately)
    res = np.empty(2*len(t))
    res[0::2] = x_pred - x_obs
    res[1::2] = y_pred - y_obs
    return res

# Helper to compute L1 score (sum of Euclidean distances)
def compute_L1_score(x_obs, y_obs, x_pred, y_pred):
    d = np.sqrt((x_obs - x_pred)**2 + (y_obs - y_pred)**2)
    return np.sum(np.abs(d))
```

```python
from scipy.optimize import minimize_scalar

# load points
df = pd.read_csv('/content/xy_data.csv')
x_obs = df['x'].to_numpy()
y_obs = df['y'].to_numpy()
N = len(x_obs)

# initial guess via linear fit (y = s*x + b)
A = np.vstack([x_obs, np.ones_like(x_obs)]).T
s, b = np.linalg.lstsq(A, y_obs, rcond=None)[0]
theta_init_deg = np.rad2deg(np.arctan(s))
X_init = (42.0 - b) / s if abs(s) > 1e-8 else 0.0
M_init = 0.0
params = np.array([theta_init_deg, M_init, X_init])

# helper: for single observed point, find best t given params
def find_best_t_for_point(xo, yo, params, t_bounds=(6.0, 60.0)):
    theta_deg, M, X = params
    theta_rad = np.deg2rad(theta_deg)
    def dist2(t):
        xp, yp = model_xy(t, theta_rad, M, X)
        return (xp - xo)**2 + (yp - yo)**2
    res = minimize_scalar(dist2, bounds=t_bounds, method='bounded', options={'xatol':1e-6})
    return res.x, res.fun

# alternating scheme
t_items = np.linspace(6,60,N)  # initial t guesses (or equally spaced)
for iteration in range(8):
    # 1) find best t_i for each observed point given current params
    for i in range(N):
        t_i, fval = find_best_t_for_point(x_obs[i], y_obs[i], params)
        t_items[i] = t_i

    # 2) optimize params given t_items (now we have paired t and observations)
    def obj_params(p):
        # p = [theta_deg, M, X]
        res = residuals_for_params_with_t(p, t_items, x_obs, y_obs)
        # return L2 norm for optimizer
        return 0.5 * np.sum(res**2)
    # bounds
    bounds = [(0.0,50.0), (-0.05,0.05), (0.0,100.0)]
    res_opt = minimize(obj_params, params, bounds=bounds, method='L-BFGS-B', options={'maxiter':200})
    params = res_opt.x
    print("Iter", iteration, "params:", params)

theta_est_deg, M_est, X_est = params
print("Final estimate:", theta_est_deg, M_est, X_est)

# compute L1 score
x_pred, y_pred = model_xy(t_items, np.deg2rad(theta_est_deg), M_est, X_est)
L1 = compute_L1_score(x_obs, y_obs, x_pred, y_pred)
print("L1 score:", L1)
```

```python
# plot observed points and fitted curve (dense)
t_dense = np.linspace(6, 60, 2000)
x_dense, y_dense = model_xy(t_dense, np.deg2rad(theta_est_deg), M_est, X_est)
plt.figure(figsize=(8,6))
plt.scatter(x_obs, y_obs, s=10, label='observed')
plt.plot(x_dense, y_dense, linewidth=1, label='fitted curve')
plt.axis('equal')
plt.legend()
plt.savefig("plot.png", dpi=300, bbox_inches='tight')
plt.show()
```
